```python
import matplotlib.pyplot as plt

# Pontos inicial e final da reta
x1, y1 = 12, 0
x2, y2 = 0, 12/(10**3)

# Ponto adicional na reta
ponto_x, ponto_y = 4.57, 7.43 * 10**-3

# Criar a figura e os eixos
fig, ax = plt.subplots()

# Desenhar a reta
ax.plot([x1, x2], [y1, y2], marker='o')

# Adicionar o ponto à reta
ax.plot(ponto_x, ponto_y, marker='o', color='red', label='Ponto de Operação 1')

ax.legend()
ax.text(ponto_x + 0.2, ponto_y, f'({ponto_x}, {ponto_y:.4f})', color='red')

# Ajustar a escala dos eixos
ax.set_xlim(0, 13)  # Definindo a escala do eixo x
ax.set_ylim(0, 0.013)  # Definindo a escala do eixo y

# Adicionar rótulos aos eixos
ax.set_xlabel('V_CE')
ax.set_ylabel('I_c')

# Adicionar título ao gráfico
ax.set_title('Reta de Carga')

# Exibir o gráfico
plt.show()
```
